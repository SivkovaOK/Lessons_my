```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sympy import *
from mpl_toolkits.mplot3d import Axes3D
```

```python
x = np.linspace(-5,5,100)
f_x = 3*(x**2)
plt.plot(x,f_x, color = 'red')
plt.grid(True)
plt.xlabel('x')
plt.ylabel('y', rotation = 0)
ax = plt.gca()
ax.axhline(y= 0, color='k')
ax.axvline(x= 0, color='k')
plt.show()
def f(x):
  return 3 *(x**2)
def df(x):
  return 6 * x
f_x = 3*(x**2)
x_0 = 5
la = 0.1
listok = []
while True:
    listok.append((x_0,f(x_0)))
    x_1 = x_0-la*df(x_0)
    if abs(x_0 - x_1) <= 1e-8:
        break
    x_0 = x_1
x_, y_ = zip(*listok)
plt.plot(x,f(x))
plt.grid(True)
plt.ylabel('y', rotation = 0)
plt.xlabel('x')
plt.plot(x_,y_,'o-', color = 'hotpink')
ax = plt.gca()
ax.axhline(y= 0, color='k')
ax.axvline(x= 0, color='k')
plt.show()
```

```python
def f(x):
  return 3 *(x[0]**2)+5*(x[1]**2)
def df(x):
  return np.array([6 * x[0] + 10*x[1]])
x_0 = np.array([5,10])
la = 0.1
li = []
epsilon = 1e-8
while True:
    li.append(np.append(x_0, f(x_0)))
    x_1 = x_0 - la * df(x_0)
    if np.linalg.norm(x_1 - x_0) <= epsilon:
        break
    x_0 = x_1
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
X = np.linspace(-5, 5, 100)
Y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(X, Y)
Z = f([X, Y])
ax.plot_surface(X, Y, Z, cmap=plt.cm.coolwarm)
ax.scatter(li[-1][0], li[-1][1], li[-1][2], color='blue', marker='o', s=50)
for i in range(len(li)-1):
    ax.plot([li[i][0], li[i+1][0]], [li[i][1], li[i+1][1]], [li[i][2], li[i+1][2]], color='blue')

plt.show()
```
